### Subarray Sum Equals K

Link - [Leetcode-560](https://leetcode.com/problems/subarray-sum-equals-k)

Approach - Prefix sums (to calc current sum) + Hashmap (to count prefix sums)

```python
from typing import List

def subarraySum(nums: List[int], k: int) -> int:
    res = 0
    prefix = 0
    prefixCount = {0:1}

    for n in nums:
        prefix += n
        diff = prefix-k
        res += prefixCount.get(diff, 0)
        prefixCount[prefix] = 1 + prefixCount.get(prefix, 0)
    return res
    
nums = [1,-1,1,1,1,1]
k = 3
subarraySum(nums, k)
```

### Continuous Subarray Sum

Link - [Leetcode-523](https://leetcode.com/problems/continuous-subarray-sum)

Approach - 
1. Hashmap({reminder: index})
2. Count prefix and find reminder -> check reminder in hashmap -> if its value is greater than 1 then its subarray multiple of k

```python
from typing import List

def checkSubarraySum(nums: List[int], k: int) -> bool:
    prefix = 0
    reminders = {0: -1} # {reminder: index}

    for i, n in enumerate(nums):
        prefix += n
        r = prefix % k

        if r not in reminders:
            reminders[r] = i
        elif i - reminders[r] > 1:
            return True
    return False

nums = [23,2,4,6,7]
k = 6
checkSubarraySum(nums, k)
```

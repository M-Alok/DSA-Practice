### 2sum

Link - [LeetCode-1](https://leetcode.com/problems/two-sum)

Approach - Iterate loop once and find composite of current value by target value

```python
from typing import List

def twoSum(nums: List[int], target: int) -> List[int]:
    count = {}

    for i in range(len(nums)):
        comp = target - nums[i]
        if comp in count:
            return [count[comp], i]
        count[nums[i]] = i
    return []

nums = [2,7,11,15]
target = 9
twoSum(nums, target)
```

### 2Sum II â€“ Input Array Is Sorted

Link - [LeetCode-167](https://leetcode.com/problems/two-sum-ii-input-array-is-sorted)

Approach - Increase or decrease pointers based on sum

```python
from typing import List

def twoSum(numbers: List[int], target: int) -> List[int]:
    l, r = 0, len(numbers)-1
    
    while l < r:
        summ = numbers[l] + numbers[r]
        if summ == target:
            return [l+1, r+1]
        elif summ < target:
            l += 1
        else:
            r -= 1

numbers = [2,7,11,15]
target = 9
twoSum(numbers, target)
```

### 3Sum

Link - [LeetCode-15](https://leetcode.com/problems/3sum)

Approach - Fix one pointer, two pointers solve inside

```python
from typing import List

def threeSum(nums: List[int]) -> List[List[int]]:
    res = []
    nums.sort()

    for i, a in enumerate(nums):
        if a > 0: break
        if i > 0 and nums[i] == nums[i-1]: continue

        l, r = i+1, len(nums)-1
        while l < r:
            threeSum = a + nums[l] + nums[r]
            if threeSum > 0:
                r -= 1
            elif threeSum < 0:
                l += 1
            else:
                res.append([a, nums[l], nums[r]])
                l += 1
                while l < r and nums[l] == nums[l-1]:
                    l += 1
    return res

nums = [-1,0,1,2,-1,-4]
threeSum(nums)
```

### 4Sum

Link - [LeetCode-18](https://leetcode.com/problems/4sum)

Approach - Same idea as 3Sum but nested loop for first 2 numbers

```python
from typing import List

def fourSum(nums: List[int], target: int) -> List[List[int]]:
    res = []
    nums.sort()

    for i in range(len(nums)):
        if i > 0 and nums[i] == nums[i-1]:
            continue

        for j in range(i+1, len(nums)-2):
            if j > i + 1 and nums[j]== nums[j-1]:
                continue
                
            l, r = j+1, len(nums)-1
            while l < r:
                summ = nums[i]+nums[j]+nums[l]+nums[r]
                if summ > target:
                    r -= 1
                elif summ < target:
                    l += 1
                else:
                    res.append([nums[i], nums[j], nums[l], nums[r]])
                    l += 1
                    while l < r and nums[l] == nums[l-1]:
                        l += 1
    return res

nums = [1,0,-1,0,-2,2]
target = 0
fourSum(nums, target)
```

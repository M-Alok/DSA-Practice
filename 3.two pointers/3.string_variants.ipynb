### Valid Palindrome

Link - [LeetCode-125](https://leetcode.com/problems/valid-palindrome)

Approach - Skip non-alphanumeric characters then compare from first to last characters

```python
def isPalindrome(s: str) -> bool:
    # Helper function to check alpha numeric of a character
    def isAlphaNum(char: str) -> bool:
        return (ord('A') <= ord(char) <= ord('Z') or
                ord('a') <= ord(char) <= ord('z') or
                ord('0') <= ord(char) <= ord('9'))

    l, r = 0, len(s)-1

    while l < r:
        while l < r and not isAlphaNum(s[l]):
            l += 1
        while l < r and not isAlphaNum(s[r]):
            r -= 1
        if s[l].lower() != s[r].lower():
            return False
        l += 1
        r -= 1
    return True

isPalindrome("A man, a plan, a canal: Panama")
```

### Valid Palindrome II

Link - [LeetCode-680](https://leetcode.com/problems/valid-palindrome-ii)

Approach - Compare from first to last characters if they mismatch then skip either first or last then compare again

```python
def validPalindrome(s: str) -> bool:
    def check(i, j):
        while i < j:
            if s[i] != s[j]:
                return False
            i, j = i+1, j-1
        return True

    l, r = 0, len(s)-1
    while l < r:
        if s[l] != s[r]:
            return check(l+1, r) or check(l, r-1)
        l, r= l+1, r-1
    return True

validPalindrome('abca')
```

### Reverse Words in a String

Link - [LeetCode-151](https://leetcode.com/problems/reverse-words-in-a-string)

Approach - Skip trialing spaces point right pointer at end of word and left pointer at start of word and append word to get result, continue it entire sentence

```python
def reverseWords(s: str) -> str:
    res = []
    l = r = len(s)-1

    while l >= 0:
        while l >= 0 and s[l] == ' ':
            l -= 1
        if l < 0:
            break

        r = l
        while l >= 0 and s[l] != ' ':
            l -= 1
        
        res .append(s[l+1:r+1])
    return ' '.join(res)

s = "   a good   example  "
reverseWords(s)
```

### Container With Most Water - 11

Link - [LeetCode-11](https://leetcode.com/problems/container-with-most-water)

Approach - Calculate area between two heights and update if it is max area that can contain water

```python
from typing import List

def maxArea(height: List[int]) -> int:
    res = 0
    l, r = 0, len(height)-1

    while l < r:
        area = min(height[l], height[r]) * (r-l)
        res = max(res, area)
        if height[l] < height[r]:
            l += 1
        else:
            r -= 1
    return res

height = [1,8,6,2,5,4,8,3,7]
maxArea(height)
```

### Longest Palindromic Substring

Link - [LeetCode-5](https://leetcode.com/problems/longest-palindromic-substring)

Approach - Find palindrome of every character by moving left and right(also compare) simultaneously

```python
def longestPalindrome(s: str) -> str:
    res = ''
    res_len = 0

    for i in range(len(s)):
        l, r = i, i
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if (r-l+1) > res_len:
                res = s[l:r+1]
                res_len = r-l+1
            l -= 1
            r += 1
        
        l, r = i, i + 1
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if (r-l+1) > res_len:
                res = s[l:r+1]
                res_len = r-l+1
            l -= 1
            r += 1
    return res

longestPalindrome("babad")
```

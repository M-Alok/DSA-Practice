### Squares of a Sorted Array

Link - [LeetCode-977](https://leetcode.com/problems/squares-of-a-sorted-array)

Approach - Two pointers moving inward from both ends

```python
from typing import List

def sortedSquares(nums: List[int]) -> List[int]:
    n = len(nums)
    res = [0] * n
    l, r = 0, n-1
    pos = n - 1

    while l <= r:
        left, right = nums[l] ** 2, nums[r] ** 2

        if left > right:
            res[pos] = left
            l += 1
        else:
            res[pos] = right
            r -= 1
        pos -= 1
    return res

nums = [-4,-1,0,3,10]
sortedSquares(nums)
```

### Reverse String

Link - [LeetCode-344](https://leetcode.com/problems/reverse-string)

Approach - Swap elements with left & right pointers

```python
from typing import List

def reverseString(s: List[str]) -> None:
    l, r = 0, len(s)-1

    while l <= r:
        s[l], s[r] = s[r], s[l]
        l, r = l + 1, r - 1

s = ["h","e","l","l","o"]
reverseString(s)
print(s)
```

### Move Zeroes

Link - [LeetCode-283](https://leetcode.com/problems/move-zeroes)

Approach - Swap non zero numbers at beginning

```python
def moveZeroes(nums: List[int]) -> None:
    non_zero = 0

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[non_zero], nums[i] = nums[i], nums[non_zero]
            non_zero += 1

nums = [0,1,0,3,12]
moveZeroes(nums)
print(nums)
```

### Remove Duplicates from Sorted Array

Link - [LeetCode-26](https://leetcode.com/problems/remove-duplicates-from-sorted-array)

Approach - Loop from 2nd element and swap if elements are different

```python
def removeDuplicates(nums: List[int]) -> int:
    j = 1
    for i in range(1, len(nums)):
        if nums[i] != nums[i - 1]:
            nums[j] = nums[i]
            j += 1
    return j

nums = [1,1,2]
removeDuplicates(nums)
```

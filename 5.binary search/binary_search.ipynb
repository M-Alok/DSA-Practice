### Binary Search

Link - [Leetcode-704](https://leetcode.com/problems/binary-search)

Approach - Find mid and shift pointers

```python
from typing import List

def search(nums: List[int], target: int) -> int:
    l, r = 0, len(nums)-1

    while l <= r:
        m = (l + r) // 2
        if target < nums[m]:
            r = m - 1
        elif target > nums[m]:
            l = m + 1
        else:
            return m
    return -1

nums = [-1,0,3,5,9,12]
target = 9
search(nums, target)
```

### Search a 2D Matrix

Link - [Leetcode-74](https://leetcode.com/problems/search-a-2d-matrix)

Approach - Find mid of 2D matrix and then apply binary search

```python
from typing import List

def searchMatrix(matrix: List[List[int]], target: int) -> bool:
    ROWS, COLS = len(matrix), len(matrix[0])
    t, b = 0, ROWS-1
    while t <= b:
        btw = (t + b) // 2
        if target < matrix[btw][0]:
            b = btw - 1
        elif target > matrix[btw][-1]:
            t = btw + 1
        else:
            break

    if not t <= b:
        return False
    
    btw = (t + b) // 2
    l, r = 0, COLS-1
    while l <= r:
        m = (l + r) // 2
        if target < matrix[btw][m]:
            r = m - 1
        elif target > matrix[btw][m]:
            l = m + 1
        else:
            return True
    return False

matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 3
searchMatrix(matrix, target)
```

### Koko Eating Bananas

Link - [Leetcode-875](https://leetcode.com/problems/koko-eating-bananas)

Approach - 
1. Apply binary search between 1 and max val in piles
2. Find total hours to eat all bananas in piles
3. If it is possible to eat all then search even min val that satisfies this condition

```python
import math
from typing import List

def minEatingSpeed(piles: List[int], h: int) -> int:
    res = float('inf')
    l, r = 1, max(piles)

    while l <= r:
        k = (l+r) // 2
        hours = 0
        for p in piles:
            hours += math.ceil(p/k)
        if hours <= h:
            res = min(res, k)
            r = k - 1
        else:
            l = k + 1
    return res

piles = [30,11,23,4,20]
h = 5
minEatingSpeed(piles, h)
```

### Find Minimum in Rotated Sorted Array

Link - [Leetcode-153](https://leetcode.com/problems/find-minimum-in-rotated-sorted-array)

Approach - 
1. Find mid val
2. Assume it as min
3. If mid larger than left most val then search right portion else search left portion

```python
from typing import List

def findMin(nums: List[int]) -> int:
    res = nums[0]
    l, r = 0, len(nums)-1

    while l <= r:
        if nums[l] < nums[r]:
            res = min(res, nums[l])
            break

        m = (l + r) // 2
        res = min(res, nums[m])
        if nums[m] >= nums[l]:
            l = m + 1
        else:
            r = m - 1
    return res

nums = [3,4,5,1,2]
findMin(nums)
```

### Search in Rotated Sorted Array

Link - [Leetcode-33](https://leetcode.com/problems/search-in-rotated-sorted-array)

Approach - 
1. Apply binary search to nums and find mid val
2. If target lies btween left most and mid values the search target in that portion else other portion
3. Simillarly if target lies btween mid and right most val search their else search in other portion
4. At some stage mid will be eauals to target then return it

```python
from typing import List

def search(nums: List[int], target: int) -> int:
    l, r = 0, len(nums)-1

    while l <= r:
        m = (l + r) // 2
        if nums[m] == target:
            return m
        
        if nums[l] <= nums[m]:
            if nums[l] <= target and target <= nums[m]:
                r = m - 1
            else:
                l = m + 1
        else:
            if nums[m] <= target and target <= nums[r]:
                l = m + 1
            else:
                r = m - 1
    return -1

# nums = [4,5,6,7,0,1,2]
nums = [4,5,0,1,2]
# target = 0
target = 1
search(nums, target)
```

### Time Based Key-Value Store

Link - [Leetcode-981](https://leetcode.com/problems/time-based-key-value-store)

Approach:
- Initialize store(hash map)
- To set value find it using key, if not their create empty and then append new value
- To get value apply binary search based on timestamp by fetching value by key

```python
class TimeMap:
    def __init__(self):
        self.store = {}
    
    def set(self, key: str, value: str, timestamp: int) -> None:
        if key not in self.store:
            self.store[key] = []
        self.store[key].append([value, timestamp])
    
    def get(self, key: str, timestamp: int) -> str:
        res = ''
        values = self.store.get(key, [])

        l, r = 0, len(values)-1
        while l <= r:
            m = (l + r) // 2
            if values[m][1] <= timestamp:
                res = values[m][0]
                l = m + 1
            else:
                r = m - 1
        return res
    
obj = TimeMap()
obj.set("foo", "bar", 1)
param_1 = obj.get("foo", 1)
param_2 = obj.get("foo", 3)
obj.set("foo", "bar2", 4)
param_3 = obj.get("foo", 4)
param_4 = obj.get("foo", 5)
print(param_1)
print(param_2)
print(param_3)
print(param_4)
```

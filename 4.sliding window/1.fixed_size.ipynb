### Max Consecutive Ones
Link - [LeetCode-485](https://leetcode.com/problems/max-consecutive-ones)

Approach - Iterate nums, increment count if ones else reset then return max count

```python
from typing import List

def findMaxConsecutiveOnes(nums: List[int]) -> int:
    count = 0
    ans = 0

    for num in nums:
        if num == 1:
            count += 1
        else:
            count = 0
        ans = max(ans, count)
    return ans

nums = [1,1,0,1,1,1]
findMaxConsecutiveOnes(nums)
```

### Maximum Subarray
Link - [LeetCode-53](https://leetcode.com/problems/maximum-subarray)

Approach - Kadane's algorithm -> init current sum to hold sum upto current element and max sum to hold max sum upto current element then loop through nums update(max of) current sum and max sum

```python
from typing import List

def maxSubArray(nums: List[int]) -> int:
    maxSum = nums[0]
    curSum = 0

    for n in nums:
        curSum = max(curSum, 0)
        curSum += n
        maxSum = max(maxSum, curSum)
    return maxSum

nums = [-2,1,-3,4,-1,2,1,-5,4]
maxSubArray(nums)
```

### Maximum Average Subarray I
Link - [LeetCode-643](https://leetcode.com/problems/maximum-average-subarray-i)

Approach - Calc avg of first k numbers and loop then nums with range k to n -> add new num(right most) and subtract old num(left most) thean calc avg and return maximum avg through out nums

```python
from typing import List

def findMaxAverage(nums: List[int], k: int) -> float:
    n = len(nums)
    cur_sum = 0

    for i in range(k):
        cur_sum += nums[i]

    max_avg = cur_sum / k

    for i in range(k, n):
        cur_sum += nums[i]
        cur_sum -= nums[i-k]

        avg = cur_sum / k
        max_avg = max(max_avg, avg)
    return max_avg

nums = [1,12,-5,-6,50,3]
k = 4
findMaxAverage(nums, k)
```

### Sliding Window Maximum
Link - [Leetcode-239](https://leetcode.com/problems/sliding-window-maximum)

Approach - 
1. Init empty queue(to add/remove first and last element) which stores index of elements in nums
2. Loop nums pop until last element in queue is larger
3. Then append current element
4. Pop left most elememt if it is smaller than windows left pointer
5. Update result array with max num(which is last element of queue)

```python
from typing import List
from collections import deque

def maxSlidingWindow(nums: List[int], k: int) -> List[int]:
    res = []
    q = deque()
    l = r = 0

    while r < len(nums):
        while q and nums[q[-1]] < nums[r]:
            q.pop()
        q.append(r)

        if l > q[0]:
            q.popleft()
        
        if (r+1) >= k:
            res.append(nums[q[0]])
            l += 1
        r += 1
    return res

nums = [1,3,-1,-3,5,3,6,7]
k = 3
maxSlidingWindow(nums, k)
```

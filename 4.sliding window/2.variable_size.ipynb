### Longest Substring Without Repeating Characters

Link - [LeetCode-3](https://leetcode.com/problems/longest-substring-without-repeating-characters)

Approach - Use hashset to keep track of characters and shrink when duplicate appears

```python
def lengthOfLongestSubstring(s: str) -> int:
    res = 0
    l = 0
    charSet = set()

    for r in range(len(s)):
        while s[r] in charSet:
            charSet.remove(s[l])
            l += 1
        charSet.add(s[r])
        res = max(res, r -l + 1)
    return res

lengthOfLongestSubstring("abcabcbb")
```

### Fruit Into Baskets

Link - [LeetCode-904](https://leetcode.com/problems/fruit-into-baskets)

Approach - Use hashmap keep track of visiting fruit and then add and remove using two left and right pointers and count total friuts and update max fruits

```python
from typing import List
from collections import defaultdict

def totalFruit(fruits: List[int]) -> int:
    total, res = 0, 0
    count = defaultdict(int)

    l = 0
    for r in range(len(fruits)):
        count[fruits[r]] += 1
        total += 1

        while len(count) > 2:
            f = fruits[l]
            count[f] -= 1
            total -= 1
            l += 1
            if not count[f]:
                count.pop(f)

        res = max(res, total)
    return res

fruits = [1,2,3,2,2]
totalFruit(fruits)
```

### Longest Repeating Character Replacement

Link - [LeetCode-424](https://leetcode.com/problems/longest-repeating-character-replacement)

Approach - Track highest frequency char

```python
def characterReplacement(s: str, k: int) -> int:
    count = {}
    maxf, res = 0, 0

    l = 0
    for r in range(len(s)):
        count[s[r]] = 1 + count.get(s[r], 0)
        maxf = max(maxf, count[s[r]])

        while (r - l + 1) - maxf > k:
            count[s[l]] -= 1
            l += 1  

        res = max(res, r - l + 1)
    return res

s = "ABAB"
k = 2
characterReplacement(s, k)  
```

### Minimum Size Subarray Sum

Link - [LeetCode-209](https://leetcode.com/problems/minimum-size-subarray-sum)

Approach - Shrink when sum â‰¥ target

```python
from typing import List

def minSubArrayLen(target: int, nums: List[int]) -> int:
    l, total = 0, 0
    res = float('inf')

    for r in range(len(nums)):
        total += nums[r]
        while total >= target:
            res = min(res, r-l+1)
            total -= nums[l]
            l += 1
    return 0 if res == float('inf') else res

target = 7
nums = [2,3,1,2,4,3]
minSubArrayLen(target, nums)
```

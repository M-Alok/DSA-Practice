### Subsets

Link - [Leetcode-78](https://leetcode.com/problems/subsets)

```python
from typing import List

def subsets(nums: List[int]) -> List[List[int]]:
    n = len(nums)
    res = []
    def dfs(i: int, cur: list):
        if i == n:
            res.append(cur[:])
            return
        
        # Including ith number
        cur.append(nums[i])
        dfs(i+1, cur)

        # Not including ith number
        cur.pop()
        dfs(i+1, cur)
    dfs(0, [])
    return res

subsets([1,2,3])
```

### Subsets II

Link - [LeetCode-90](https://leetcode.com/problems/subsets-ii)

Approach - Sort and skip adjacent duplicates

```python
def subsetsWithDup(nums: List[int]) -> List[List[int]]:
    nums.sort()
    n = len(nums)
    res = []

    def dfs(i: int, cur: list):
        if i == n:
            res.append(cur[:])
            return
        
        cur.append(nums[i])
        dfs(i+1, cur)
        cur.pop()
        while i + 1 < n and nums[i] == nums[i+1]:
            i += 1
        dfs(i+1, cur)
    dfs(0, [])
    return res

subsetsWithDup([1,2,2])
```

### Combination Sum

Link - [Leetcode-39](https://leetcode.com/problems/combination-sum)

```python
def combinationSum(cand: List[int], target: int) -> List[List[int]]:
    n = len(cand)
    res = []
    def dfs(i: int, cur: list, total: int) -> None:
        if total == target:
            res.append(cur[:])
            return
        if i == n or total > target:
            return
        
        cur.append(cand[i])
        dfs(i, cur, total+cand[i])  # Including same number(i)

        cur.pop()
        dfs(i+1, cur, total)
    
    dfs(0, [], 0)
    return res

candidates = [2,3,6,7]
target = 7
combinationSum(candidates, target)
```

### Combination Sum II

Link - [Leetcode-40](https://leetcode.com/problems/combination-sum-ii)

Approach - Sort and skip adjacent duplicates

```python
def combinationSum2(candidates: List[int], target: int) -> List[List[int]]:
    candidates.sort()
    n = len(candidates)
    res = []
    def dfs(i: int, cur: list, total: int) -> None:
        if total == target:
            res.append(cur[:])
            return
        if i == n or total > target:
            return
        
        cur.append(candidates[i])
        dfs(i+1, cur, total+candidates[i])

        cur.pop()
        while i + 1 < n and candidates[i] == candidates[i+1]:   # Skiping same elements
            i += 1
        dfs(i+1, cur, total)
    dfs(0, [], 0)
    return res

candidates = [10,1,2,7,6,1,5]
target = 8
combinationSum2(candidates, target)
```

### Permutations

Link - [Leetcode-46](https://leetcode.com/problems/permutations/description)

```python
from typing import List

def permute(nums: List[int]) -> List[List[int]]:
    res = []
    
    def dfs(cur: list):
        if len(cur) == len(nums):
            res.append(cur[:])
            return 
        
        for num in nums:
            if num not in cur:
                cur.append(num)
                dfs(cur)
                cur.pop()
    dfs([])
    return res

permute([1,2,3])
```

### Combinations

Link - [Leetcode-77](https://leetcode.com/problems/combinations)

```python
def combine(n: int, k: int) -> List[List[int]]:
    res = []

    def dfs(i: int, cur: list):
        if len(cur) == k:
            res.append(cur[:])
            return
        
        for num in range(i, n+1):
            cur.append(num)
            dfs(num+1, cur)
            cur.pop()
    dfs(1, [])
    return res

combine(4, 2)
```

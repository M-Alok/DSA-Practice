### Valid Parentheses

Link - [Leetcode-20](https://leetcode.com/problems/valid-parentheses)

Approach:
- Make hash map and push only openbrackets and compre closing brackeks to top of the stack
- If their is mismatch then return else true if order of every brackets is correct

```python
def isValid(s: str) -> bool:
    stack = []
    pairs = {')': '(', ']': '[', '}': '{'}

    for ch in s:
        if ch in pairs:
            if stack and stack[-1] == pairs[ch]:
                stack.pop()
            else:
                return False
        else:
            stack.append(ch)
    return not stack

isValid("([])")
```

### Evaluate Reverse Polish Notation

Link - [Leetcode-150](https://leetcode.com/problems/evaluate-reverse-polish-notation)

Approach: Iterate tokens is its number then push to stack else preform operation on previous 2 operands by poping them

```python
from typing import List

def evalRPN(tokens: List[str]) -> int:
    stack = []
    for c in tokens:
        if c == '+':
            stack.append(stack.pop() + stack.pop())
        elif c == '-':
            a = stack.pop()
            b = stack.pop()
            stack.append(b - a)
        elif c == '*':
            stack.append(stack.pop() * stack.pop())
        elif c == '/':
            a = stack.pop()
            b = stack.pop()
            stack.append(int(b / a) if b * a >= 0 else -(abs(b) // abs(a)))
        else:
            stack.append(int(c))
    return stack[0]

tokens = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
evalRPN(tokens)
```

### Daily Temperatures

Link - [Leetcode-739](https://leetcode.com/problems/daily-temperatures)

Approach: 
- Make empty list of length temperatures
- Then enumerate temperatures and push temperature with index to the stack
- Pop untile current temperature is smaller than last temperature in stack
- While poping update res by subtracting indexes(current and popped)

```python
from typing import List

def dailyTemperatures(temperatures: List[int]) -> List[int]:
    stack = []
    res = [0] * len(temperatures)
    for i, temp in enumerate(temperatures):
        while stack and temp > stack[-1][0]:
            _, idx = stack.pop()
            res[idx] = i - idx
        stack.append([temp, i])
    return res

temperatures = [73,74,75,71,69,72,76,73]
dailyTemperatures(temperatures)
```

### Car Fleet

Link - [Leetcode-853](https://leetcode.com/problems/car-fleet)

Approach:
- Zip position and speed
- Iterate reverse sorted array
- Calculate time and push it to stack
- If last speed is smaller than last but one speed pop it(because it catches smaller time car)
- At the end length of stack id number of car fleet

```python
from typing import List

def carFleet(target: int, position: List[int], speed: List[int]) -> int:
    pairs = [[p, s] for p, s in zip(position, speed)]
    stack = []

    for p, s in sorted(pairs)[::-1]:
        stack.append((target - p) / s)
        if len(stack) >= 2 and stack[-1] <= stack[-2]:
            stack.pop()
    return len(stack)

target = 12
position = [10,8,0,5,3]
speed = [2,4,1,1,3]
carFleet(target, position, speed)
```

### Largest Rectangle in Histogram

Link - [Leetcode-84](https://leetcode.com/problems/largest-rectangle-in-histogram)

Approach:
- Use stack to track previous height in histogram
- If height of smaller that previous height the pop previous height and index
- Compute max area and update index with previous index(because it can be extended backwards)
- Then compute max area of remaining elements in the stack(which can extend upto end of histogram)

```python
from typing import List

def largestRectangleArea(heights: List[int]) -> int:
    maxArea = 0
    stack = []

    for i, h in enumerate(heights):
        start = i
        while stack and stack[-1][1] > h:
            idx, height = stack.pop()
            maxArea = max(maxArea, height * (i - idx))
            start = idx
        stack.append((start, h))

    for i, h in stack:
        maxArea = max(maxArea, h * (len(heights) - i))
        
    return maxArea

heights = [2,1,5,6,2,3]
largestRectangleArea(heights)
```

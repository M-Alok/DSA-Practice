### Running Sum of 1D Array

Link - [Leetcode-1480](https://leetcode.com/problems/running-sum-of-1d-array)

Approach - Compute sum by itrating nums

```python
from typing import List

def runningSum(nums: List[int]) -> List[int]:
    prefix = [0] * len(nums)
    prefix[0] = nums[0]

    for i in range(1, len(nums)):
        prefix[i] = prefix[i-1] + nums[i]
    return prefix

nums = [1,2,3,4]
runningSum(nums)
```

### Find Pivot Index

Link - [LeetCode-724](https://leetcode.com/problems/find-pivot-index)

Approach - Prefix + total sum comparison

```python
from typing import List

def pivotIndex(nums: List[int]) -> int:
    leftSum, rightSum = 0, sum(nums)

    for i, num in enumerate(nums):
        rightSum -= num
        if rightSum == leftSum:
            return i
        leftSum += num
    return -1

nums = [1,7,3,6,5,6]
pivotIndex(nums)
```

### Range Sum Query - Immutable

Link - [Leetcode-303](https://leetcode.com/problems/range-sum-query-immutable)

Approach - Precompute prefix sum array and suntract left-1 prefix sum to right prefix sum to get cumulative sum

```python
from typing import List

class NumArray:
    def __init__(self, nums: List[int]):
        self.prefix = []
        cur = 0
        for n in nums:
            cur += n
            self.prefix.append(cur)

    def sumRange(self, left: int, right: int) -> int:
        rightSum = self.prefix[right]
        leftSum = self.prefix[left - 1] if left > 0 else 0
        return rightSum - leftSum

nums = [-2, 0, 3, -5, 2, -1]
obj = NumArray(nums)
param_1 = obj.sumRange(1, 3)
print(param_1)
```

### Minimum Value to Get Positive Step by Step Sum

Link - [Leetcode-1413](https://leetcode.com/problems/minimum-value-to-get-positive-step-by-step-sum)

Approach - Calculate min value while running sum -> return make min value +ve and increment it to get start value

```python
from typing import List

def minStartValue(nums: List[int]) -> int:
    prefix = 0
    minPrefix = float('inf')

    for n in nums:
        prefix += n
        minPrefix = min(minPrefix, prefix)
    
    return max(1, 1-minPrefix)

nums = [-3,2,-3,4,2]
minStartValue(nums)
```
